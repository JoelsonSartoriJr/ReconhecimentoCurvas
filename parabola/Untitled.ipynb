```python
from matplotlib.patches import Ellipse
import matplotlib as mpl
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import random
import pandas as pd
import glob
```

```python
xi = np.random.uniform(low=0, high=50, size=(50,))
xf = np.random.uniform(low=0, high=50, size=(50,))
data = {'xi':xi, 'xf':xf}
df = pd.DataFrame(data)
df.to_csv("ParabolaYnegSemGrid.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def parabola(xi, xf, cont):
    t = np.arange(-xi, xf, 0.01)
    x= -t**2 + -2*t
    y=t + 1

    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.autoscale()
    #ax.grid()
    plt.savefig("{}parabola.png".format(cont))
    plt.show()
```

```python
cont = 250
for d in df.values:
    parabola(d[0],d[1], cont)
    cont +=1
```
