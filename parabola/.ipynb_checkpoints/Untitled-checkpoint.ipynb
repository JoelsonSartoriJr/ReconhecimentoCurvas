```python
from matplotlib.patches import Ellipse
import matplotlib as mpl
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import random
import pandas as pd
import glob
```

```python
xi = np.random.uniform(low=0, high=100, size=(100,))
xf = np.random.uniform(low=-100, high=0, size=(100,))
data = {'xi':xi, 'xf':xf}
df = pd.DataFrame(data)
df.to_csv("SemFundoGrid.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def parabola(xi, xf):
    t = np.arange(-xi, xf, 0.01)
    x= t**2 + -2*t
    y=t + 1

    fig, ax = plt.subplots()
    ax.plot(x, y)

    ax.grid(True, linestyle='-.')

    plt.show()
```

```python
for d in df.values:
    parabola(d[0], d[1])
```
