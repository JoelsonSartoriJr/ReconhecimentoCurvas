```python
from matplotlib.patches import Ellipse
import matplotlib as mpl
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import random
import pandas as pd
import glob
```

```python
mean1 = np.random.uniform(low=-180, high=180, size=(200,))
mean2 = np.random.uniform(low=-180, high=180, size=(200,))
width = np.random.uniform(low=-180, high=180, size=(200,))
height = np.random.uniform(low=-180, high=180, size=(200,))
angle = np.random.uniform(low=0, high=180, size=(200,))
```

```python
data = {'mean1':mean1, 'mean2':mean2, 'width':width, 'height':height, 'angle':angle}
```

```python
df = pd.DataFrame(data)
df.to_csv("SemFundo.csv")
```

```python
def elipse(mean, width, height, angle, cont):
    ell = mpl.patches.Ellipse(xy=mean, width=width, height=height, angle = 180+angle, facecolor='w', edgecolor='grey')
    fig, ax = plt.subplots()

    ax.add_patch(ell)
    ax.set_aspect('equal')
    ax.autoscale()
    ax.set_axis_off()
    #ax.grid()
    plt.savefig("elipse.{}.png".format(cont))
    plt.show()
```

```python
cont = 200
for d in df.values:
    elipse([d[0],d[1]], d[2], d[3], d[4], cont)
    cont += 1
```

```python
def createTest(data, nameFile):
    file = open(nameFile, 'w')
    for i in glob.glob(str(data)+'*.png'):
        file.writelines(i+',elipse'+'\n')
    print(nameFile + ' criado')
    file.close()
    
createTest('', 'elipse.csv')
```
