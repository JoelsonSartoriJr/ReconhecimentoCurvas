```python
from matplotlib.patches import Ellipse
import matplotlib as mpl
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import random
import pandas as pd
import glob
```

```python
xi = np.random.uniform(low=0, high=100, size=(100,))
xf = np.random.uniform(low=0, high=100, size=(100,))
data = {'xi':xi, 'xf':xf}
df = pd.DataFrame(data)
df.to_csv("parabolaYpos.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def parabola(xi, xf, cont):
    t = np.arange(-xi, xf, 0.01)
    x= t**2 + -2*t
    y=t + 1

    fig, ax = plt.subplots()
    ax.plot(y, x)
    ax.autoscale()
    ax.set_axis_off()
    #ax.grid()
    plt.savefig("parabola.{}.png".format(cont))
    plt.show()
```

```python
cont = 300
for d in df.values:
    parabola(d[0],d[1], cont)
    cont +=1
```

```python
def createTest(data, nameFile):
    file = open(nameFile, 'w')
    for i in glob.glob(str(data)+'*.png'):
        file.writelines(i+',parabola'+'\n')
    print(nameFile + ' criado')
    file.close()
    
createTest('', 'parabola.csv')
```
