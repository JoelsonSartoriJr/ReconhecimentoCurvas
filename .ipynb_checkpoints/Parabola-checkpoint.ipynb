```python
import matplotlib.pyplot as plt
import numpy as np
import random
import pandas as pd
```

```python
def parabola(x,a,b,c):
    y =  a*x**2 + b*x + c
    x = np.linspace(-x, x, 2*x)
    fig, ax = plt.subplots()
    ax.plot(x, y)
```

```python
a = random.sample(range(-10,10), 20)
b = random.sample(range(-10,10), 20)
c = random.sample(range(-10,10), 20)
x = random.sample(range(-10,10), 20)
y = [a*x**2 + b*x + c for a,b,c,x in zip(a,b,c,x)]
```

```python
data = {'a':a, 'b':b, 'c':c, 'x':x, 'y':y}
```

```python
df = pd.DataFrame(data)
```
