```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("static.csv")
df
```

```python
plt.figure(figsize=(20,12),dpi=96) #96 is my screen DPI
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = df["rede"][:4]
y_pos = np.arange(len(objects))
performance = df["meanAcc"][:4]*100
 
plt.bar(y_pos, performance, align='center', alpha=0.5, color=("purple", "slategrey", "darkolivegreen", "teal"))
plt.xticks(y_pos, objects)
plt.ylabel('Mean')
plt.title('Mean acurracy with 1000 epochs')
plt.grid(True)
plt.savefig("MeanAccuracy1000epochs",type="png",dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = df["rede"][4:]
y_pos = np.arange(len(objects))
performance = df["meanAcc"][4:]*100
 
plt.bar(y_pos, performance, align='center', alpha=0.5, color=("purple", "slategrey", "darkolivegreen", "teal"))
plt.xticks(y_pos, objects)
plt.ylabel('Mean')
plt.title('Mean acurracy with 3000 epochs')
plt.grid(True) 
plt.savefig("MeanAccuracyur3000epochs",type="png",dpi=300)
plt.show()
```
